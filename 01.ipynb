# Part 1

```python
rotations = ["L68",
             "L30",
             "R48",
             "L5",
             "R60",
             "L55",
             "L1",
             "L99",
             "R14",
             "L82"]
```

```python
def applyRotation(rotation: str, state: int):
    direction = rotation[:1]
    n = int(rotation[1:])
    
    if direction == "L":
        n *= -1
    
    return (state + n) % 100

def getPassword(rotations: list[str]):
    state = 50
    count = 0

    for r in rotations:
        state = applyRotation(r, state)
                
        if state == 0:
            count += 1
    
    return count

getPassword(rotations)
```

```python
with open("01-input.txt") as f:
    input = [line.strip() for line in f]

getPassword(input)
```

# Part 2

```python
def applyRotation(rotation: str, state: tuple[int, int]):
    direction = rotation[:1]
    n = int(rotation[1:])

    if direction == "L":
        n *= -1

    x = state[0] + n
    remainder = x % 100

    s2 = abs(int(x / 100)) + state[1]
    if (x <= 0 and state[0] != 0):
        s2 += 1

    return (remainder, s2)

def getPassword(rotations: list[str]):
    state = (50, 0)

    for r in rotations:
        state = applyRotation(r, state)
    
    return state

getPassword(rotations)
```

```python
with open("01-input.txt") as f:
    input = [line.strip() for line in f]

getPassword(input)
```
