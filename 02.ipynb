```python
def getInvalidIds(input: str):
    ranges = [rangeStr.split("-") for rangeStr in input.split(",")]
    
    ids = []
    for r in ranges:
        start = int(r[0])
        end = int(r[1])

        for n in range(start, end + 1):
            nStr = str(n)
            isInvalid = False

            indexMiddle = int(len(nStr) / 2)
            left = nStr[:indexMiddle]
            right = nStr[indexMiddle:]

            if left == right:
                ids.append(n)
    
    return ids

ids = getInvalidIds(input)
print(ids)
print(sum(ids))


```
