```python
testInput = """
987654321111111
811111111111119
234234234234278
818181911112111
""".strip()
```

```python
def getJoltages(input: str):
    joltages = []
    for bank in input.splitlines():
        d1 = 0
        d2 = 0
        for i in range(len(bank) - 1):
            x = int(bank[i])
            y = int(bank[i + 1])

            if x > d1:
                d1 = x
                d2 = y
            elif y > d2:
                d2 = y

        n = int(str(d1) + str(d2))
        joltages.append(n)

    return joltages

joltages = getJoltages(testInput)
print(joltages)
print(sum(joltages))
```

```python
with open("03-input.txt") as f:
    input = f.read()

joltages = getJoltages(input)
print(joltages)
print(sum(joltages))
```
