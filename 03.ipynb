# Part 1

```python
testInput = """
987654321111111
811111111111119
234234234234278
818181911112111
""".strip()
```

```python
def getJoltages(input: str):
    joltages = []
    for bank in input.splitlines():
        d1 = 0
        d2 = 0
        for i in range(len(bank) - 1):
            x = int(bank[i])
            y = int(bank[i + 1])

            if x > d1:
                d1 = x
                d2 = y
            elif y > d2:
                d2 = y

        n = int(str(d1) + str(d2))
        joltages.append(n)

    return joltages

joltages = getJoltages(testInput)
print(joltages)
print(sum(joltages))
```

```python
with open("03-input.txt") as f:
    input = f.read()

joltages = getJoltages(input)
print(joltages)
print(sum(joltages))
```

# Part 2

```python
N_DIGITS = 12

def getJoltagesP2(input: str):
    joltages = []
    for bank in input.splitlines():
        print(bank)
        digits = [0] * N_DIGITS
        for i in range(len(bank) - N_DIGITS):
            for j in range(N_DIGITS):
                index = i + j
                if int(bank[index]) > digits[j]:
                    digits = digits[:j] + [int(d) for d in bank[index:i+N_DIGITS]]
                    print(digits)
        n = int("".join([str(d) for d in digits]))
        joltages.append(n)

    return joltages

joltages = getJoltagesP2(testInput)
print(joltages)
print(sum(joltages))
```
