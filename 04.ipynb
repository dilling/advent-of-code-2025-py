```python
testInput = """
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
""".strip()
```

```python
def checkLocation(arr: list[list[str]], x: int, y: int) -> bool:
    total = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i==0 and j==0:
                continue

            locX = x + i
            locY = y+j

            if locX >= len(arr) or locX < 0 or locY >= len(arr[locX]) or locY < 0:
                continue

            loc = arr[x+i][y+j]
                
            if loc == "@":
                total += 1

            if total >= 4:
                return False


    return True
            

def checkRolls(input: str):
    arr = [list(i) for i in input.splitlines()]
    total = 0
    testArr = [[True] * len(arr[0]) for _ in range(len(arr))]

    for x in range(len(arr)):
        for y in range(len(arr[x])):
            if arr[x][y] != '@':
                continue
            
            isValid = checkLocation(arr, x, y)

            testArr[x][y] = isValid

            if isValid:
                total += 1

    for row in testArr:
        print(row)

    return total
            

checkRolls(testInput)
```
