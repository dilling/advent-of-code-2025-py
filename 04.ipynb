# Part 1

```python
testInput = """
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
""".strip()
```

```python
def checkLocation(arr: list[list[str]], x: int, y: int) -> bool:
    total = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i==0 and j==0:
                continue

            locX = x + i
            locY = y+j

            if locX >= len(arr) or locX < 0 or locY >= len(arr[locX]) or locY < 0:
                continue

            loc = arr[x+i][y+j]
                
            if loc == "@":
                total += 1

            if total >= 4:
                return False


    return True
            

def checkRolls(input: str):
    arr = [list(i) for i in input.splitlines()]
    total = 0

    for x in range(len(arr)):
        for y in range(len(arr[x])):
            if arr[x][y] != '@':
                continue
            
            isValid = checkLocation(arr, x, y)

            if isValid:
                total += 1
                
    return total
            

checkRolls(testInput)
```

```python
with open("04-input.txt") as f:
    realInput = f.read()

checkRolls(realInput)
```

# Part 2

```python
def checkRollsP2(input: str):
    arr = [list(i) for i in input.splitlines()]
    total = 0

    complete = False
    while not complete:
        complete = True
        for x in range(len(arr)):
            for y in range(len(arr[x])):
                if arr[x][y] != '@':
                    continue
                
                isValid = checkLocation(arr, x, y)

                if isValid:
                    total += 1
                    arr[x][y] = '.'
                    complete = False
                
    return total

checkRollsP2(testInput)
```

```python
with open("04-input.txt") as f:
    realInput = f.read()

checkRollsP2(realInput)
```
