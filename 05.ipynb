```python
testInput = """
3-5
10-14
16-20
12-18

1
5
8
11
17
32
""".strip()
```

```python
def parseAndFormatRanges(rangeStr: str):
    ranges = [(int(start), int(end)) for (start, _, end) in map(lambda r: r.partition("-"), rangeStr.splitlines())]
    ranges = sorted(ranges, key=lambda r: r[0])

    reducedRanges = ranges[:1]
    for r in ranges[1:]:
        prev = reducedRanges[-1]
        if prev[1] > r[0]:
            reducedRanges[-1] = (prev[0], r[1])
        else:
            reducedRanges.append(r)

    return reducedRanges

def getFreshIngredients(input: str):
    rangeStrings, _, ids = input.partition("\n\n")
    freshIds = []
    ids = [int(id) for id in ids.splitlines()]
    print(len(ids))
    
    ranges = parseAndFormatRanges(rangeStrings)
    for id in ids:
        for (start, end) in ranges:
            if id >= start and id <= end:
                freshIds.append(id)
                break

    return freshIds



freshIngredients = getFreshIngredients(testInput)
print(freshIngredients)
print(len(freshIngredients))
```
