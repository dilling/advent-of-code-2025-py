# Part 1

```python
testInput = """
123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
""".strip("\n")
```

```python
import re 

def parseInput(input: str):
    lines = input.splitlines()
    operations = re.split(r"\s+", lines[-1].strip())
    problems = [{ "n": [], "op": op} for op in operations]

    for i in range(len(lines[:-1])):
        cells = [int(x) for x in re.split(r"\s+", lines[i].strip())]
        for j in range(len(cells)):
            problems[j]["n"].append(cells[j])

    return(problems)

parseInput(testInput)
```

```python
def doMath(input: str):
    problems = parseInput(input)
    for p in problems:
        ans = p["n"][-1]
        for n in p["n"][:-1]:
            if p["op"] == "+":
                ans += n
            elif p["op"] == "*":
                ans *= n

        p["ans"] = ans

    return problems

res = doMath(testInput)
print(res)
print(sum([p["ans"] for p in res]))

```

```python
with open("06-input.txt") as f:
    realInput = f.read()

res = doMath(realInput)
print(res)
print(sum([p["ans"] for p in res]))
```

```python
import re 

def parseInput(input: str):
    lines = input.splitlines()
    newLines = [[] for _ in lines[0]]

    for i in range(len(newLines)):
        for j in range(len(lines)):
            newLines[i].append(lines[j][i])

    newLines = ["".join(l) for l in newLines]
    print(newLines)

    problems = []
    for line in newLines:
        if line.strip() == '':
            continue

        num = int(line[:-1].strip())
        op = line[-1]

        if op != ' ':
            problems.append({ "n": [num], "op": op})
        else:
            problems[-1]["n"].append(num)
    
    return problems

parseInput(testInput)
```

```python
with open("06-input.txt") as f:
    realInput = f.read()

res = doMath(realInput)
print(res)
print(sum([p["ans"] for p in res]))
```
