```python
testInput = """
.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
""".strip("\n")
```

```python
def parseInput(input: str):
    return [list(line) for line in input.splitlines()]

parseInput(testInput)
```

```python
def calculateTachyons(input: str):
    manifold = parseInput(input)
    height = len(manifold)
    width = len(manifold[0])
    numSplit = 0

    for y in range(1, height):
        for x in range(width):
            aboveCell = manifold[y-1][x]
            if not (aboveCell == "|" or aboveCell == "S"):
                continue

            cell = manifold[y][x]
            if cell == ".":
                manifold[y][x] = "|"
            elif cell == "^":
                manifold[y][x-1] = "|"
                manifold[y][x+1] = "|"
                numSplit += 1

    print(numSplit)

    return "\n".join(["".join(c) for c in manifold])

manifold = calculateTachyons(testInput)
print(manifold)    
```

```python
with open("07-input.txt") as f:
    realInput = f.read()

manifold = calculateTachyons(realInput)
```
